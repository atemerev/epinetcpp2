```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
plt.rcParams['figure.figsize'] = [10, 7]
```

```python
df = pd.read_csv("out.csv", header=None)
df
```

```python
plt.figure()
plt.grid(True)
plt.title("Infection cases by day")
plt.xlabel("Time, days")
plt.ylabel("Cases count")
plt.plot(df[0], df[1])
plt.savefig("mine-cases.pdf")
```

```python
plt.figure()
plt.grid(True)
plt.title("Total infected count")
plt.xlabel("Time, days")
plt.ylabel("Infected count")
plt.plot(df[0], df[2])
plt.savefig("mine-total.pdf")
```

```python
plt.figure()
plt.grid(True)
plt.title("Mean population susceptibility")
plt.xlabel("Time, days")
plt.ylabel("Mean susceptibility")
plt.plot(df[0], df[3])
```
